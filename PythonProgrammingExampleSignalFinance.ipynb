## import pandas

```python
import pandas as pd
```

## Add MACD column to dataframe and remove NaN row

```python
def addMACD(df: pd.DataFrame,
            short_win:int=12,
            long_win:int=26) -> pd.DataFrame:
  close_price = df['Close'].shift(1)
  ma_short = close_price.rolling(short_win).mean()
  ma_long = close_price.rolling(long_win).mean()
  macd = ma_short - ma_long
  df['MACD'] = macd
  return df.dropna()
```

## example

```python
btc_with_macd = addMACD(btc)
btc_with_macd.head()
```

## MACD Trade with Long-Short

```python
def tradeUsingMACD(df: pd.DataFrame,
                   initial_wealth: float=1_000_000):
  df_with_macd = addMACD(df) # add macd column
  security = 0 # no securities
  cash = initial_wealth  # all cash in money
  wealth = pd.Series()
  for date in df_with_macd.index: # each day
    current_macd = df_with_macd.MACD[date]
    current_price = df_with_macd.Close[date]
    current_macd_precentage = current_macd / current_price
    if (current_macd_precentage > 0.00) and (security <= 0): # long
      cash_short = security * current_price  #close short
      cash = cash + cash_short
      security = cash / current_price
      cash = 0.0
    elif (current_macd_precentage < -0.00) and (security > 0): # short-sell
      cash = security * current_price
      security = 0.0
      security = -cash / current_price
      cash = 2 * cash
    else:
      pass
    current_wealth = cash + security * current_price
    wealth[date]  = current_wealth
  return wealth
```

## MACD Trade with Long Only

```python
def tradeUsingMACDLongOnly(df: pd.DataFrame,
                           initial_wealth: float=1_000_000):
  df_with_macd = addMACD(df) # add macd column
  security = 0 # no securities
  cash = initial_wealth  # all cash in money
  wealth = pd.Series()
  for date in df_with_macd.index: # each day
    current_macd = df_with_macd.MACD[date]
    current_price = df_with_macd.Close[date]
    current_macd_precentage = current_macd / current_price
    if (current_macd_precentage > 0.00) and (security == 0): # buy
      security = cash / current_price
      cash = 0.0
    elif (current_macd_precentage < -0.00) and (security > 0): # sell
      cash = security * current_price
      security = 0.0
    else:
      pass
    current_wealth = cash + security * current_price
    wealth[date] = current_wealth
  return wealth
```

## Buy and Hold

```python
def TradeWithBuyAndHold(df:pd.DataFrame,
                        initial_wealth: float=1_000_000):
  security = 0 # no securities
  cash = initial_wealth  # all cash in money
  wealth = pd.Series()
  for cnt, date in enumerate(df.index):
    current_price = df.Close[date]
    if cnt == 0: # buy
      security = cash / current_price
      cash = 0.0
    current_wealth = cash + security * current_price
    wealth[date]  = current_wealth
  return wealth
```

## Compare all

```python
wealthMACDLongShort = tradeUsingMACD(btc)
wealthMACDLongOnly = tradeUsingMACDLongOnly(btc)
wealthBH = TradeWithBuyAndHold(btc)
```

```python
plt.figure(figsize=(15,10))
plt.plot(wealthMACDLongShort/ 1_000_000 -1, label="MACD Long-Short")
plt.plot(wealthMACDLongOnly/ 1_000_000 -1, label="MACD Long Only")
plt.plot(wealthBH/ 1_000_000 -1 , label="Buy and Hold")
plt.legend()
plt.title("Performance Comparison")
plt.xlabel("Date")
plt.ylabel("Return")
plt.grid()
```
