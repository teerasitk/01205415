# import libraries

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Assign stock list

```python
stock_list = ['ADVANC', 'AOT', 'AWC', 'BANPU', 'BBL', 'BDMS', 'BEM', 'BGRIM', 'BH','BLA', 'BTS', 'CBG', 'CPALL', 'CPF', 'CPN', 'CRC', 'EA', 'EGCO', 'GLOBAL', 'GPSC', 'GULF', 'HMPRO', 'INTUCH', 'IRPC', 'IVL', 'JMART', 'JMT', 'KBANK', 'KCE', 'KTB', 'KTC', 'LH', 'MINT', 'MTC','OR', 'OSP', 'PTT', 'PTTEP', 'PTTGC', 'SAWAD', 'SCC', 'SCGP','TIDLOR', 'TISCO', 'TOP', 'TRUE', 'TTB', 'TU']
```

# compute Q1 & Q2 daily log return

```python
return_Q1Q2_2023 = {}
for sym in stock_list:
  df = yf.download(sym + ".BK", "2023-01-01", "2023-06-30")
  adj_close = df["Adj Close"]
  log_ret = np.log(adj_close.pct_change(1)+1)
  log_ret = log_ret.dropna()
  return_Q1Q2_2023[sym] = log_ret
return_Q1Q2_2023 = pd.DataFrame(return_Q1Q2_2023)
return_Q1Q2_2023 = return_Q1Q2_2023.dropna()
return_Q1Q2_2023.head()
```

# K-mean clustering
A good investor always diversifies his or her portfolio. Assume that you want to invest on five stocks from stock list

1.	Using K-Mean algorithm to divide the stock into 5 groups based on their daily log-return in Q1 & Q1 2023


```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=12345)
kmeans.fit(return_Q1Q2_2023.values.T) # keams to fit
for k in range(5):
  idx = np.where(kmeans.labels_ == k)[0] # find stock in the cluster
  txt = f"Cluster {k+ 1} contains"
  for idk in idx:
    txt += f" {return_Q1Q2_2023.columns[idk]}, "
  print(txt)
```

# My criteria to pick stocks that have highest return over the risk (STD) in the group

```python
# code is here
for k in range(5):
  idx = np.where(kmeans.labels_ == k)[0]
  group_sym = [stock_list[k] for k in idx]
  mean_return = return_Q1Q2_2023[group_sym].mean() # find average return
  std_return = return_Q1Q2_2023[group_sym].std() # find average std
  ratio = mean_return / std_return
  print("Best of the groups are:")
  print(ratio[ratio==ratio.max()]) # find the maximum one
```

```python
selection_list = ["KTB", "TRUE", "JMT", "TTB", "MTC"]
CASH  = 100_000
profit_seletion_list = {}
for sym in selection_list:
  df = yf.download(sym + ".BK", "2023-07-01", "2023-12-31")
  close_price = df["Close"]
  shares = CASH /  close_price.iloc[0]
  val = shares * close_price.iloc[-1]
  profit = val - CASH
  profit_seletion_list[sym] = profit
profit_seletion_list = pd.Series(profit_seletion_list)
print("\n",profit_seletion_list)
print(f"Total Gain/ loss: {profit_seletion_list.sum():,}")
```
